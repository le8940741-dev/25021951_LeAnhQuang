```python
#W2A3

# Nhập hai số nguyên từ bàn phím
a = int(input("Nhập số nguyên a: "))
b = int(input("Nhập số nguyên b: "))

# Thực hiện các phép toán
tong = a + b
hieu = a - b
tich = a * b
phan_nguyen = a // b     # Chia lấy phần nguyên
phan_du = a % b          # Chia lấy phần dư
chia_thuc = a / b        # Chia thực

# In kết quả
print("Tổng:", tong)
print("Hiệu:", hieu)
print("Tích:", tich)
print("Phần nguyên:", phan_nguyen)
print("Phần dư:", phan_du)
print("Chia thực: {:.2f}".format(chia_thuc))  # Lấy 2 chữ số thập phân

```

```python
#W2A4

# Nhập 6 đầu điểm trên một dòng, cách nhau bởi khoảng trắng
a1, b1, c1, a2, b2, a3 = map(float, input().split())

# Tính điểm trung bình theo công thức
TB = ((a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3) / 10

# In ra kết quả với 1 chữ số thập phân
print("{:.1f}".format(TB))
```

```python
#W2A8

# Nhập nhiệt độ Celsius từ bàn phím
celsius = float(input("Nhập nhiệt độ (°C): "))

# Công thức chuyển đổi sang Fahrenheit
fahrenheit = (9/5) * celsius + 32

# In kết quả với 2 chữ số thập phân
print(f"Nhiệt độ tương ứng: {fahrenheit:.2f} °F")


```

```python
#W2A9

# Nhập giá chiếc đồng hồ (USD)
x = float(input("Nhập giá đồng hồ Movado (USD): "))

# Phí vận chuyển từ nước trung gian
shipping_fee = 10  

# Thuế xuất nhập cảnh (30%)
tax_import = 0.30 * x  

# Thuế VAT (10%)
tax_vat = 0.10 * x  

# Tổng số tiền phải trả
total = x + shipping_fee + tax_import + tax_vat

# In kết quả với 2 chữ số thập phân
print(f"Tổng số tiền Đạt phải trả: {total:.2f} USD")
```

```python
#W2A10

# Nhập tên 3 người trên 1 dòng
names = input("Nhập tên 3 người (cách nhau bởi khoảng trắng): ")

# Tách tên thành danh sách
name_list = names.split()

# Đảo ngược thứ tự
name_list.reverse()

# Ghép vào chuỗi theo định dạng yêu cầu
print(f"Hi {name_list[0]}, {name_list[1]} and {name_list[2]}.")
```

```python
#W2A11

# Nhập số giờ và số phút
h = int(input("Nhập số giờ: "))
m = int(input("Nhập số phút: "))

# Chuyển đổi sang giây
seconds = h * 3600 + m * 60

# In kết quả
print(f"Tổng thời gian = {seconds} giây")
```

```python
#W2A13

# Nhập hai số nguyên dương
a = int(input("Nhập số a: "))
b = int(input("Nhập số b: "))

# Tính tích
product = a * b

# Lấy hàng đơn vị (dùng phép chia lấy dư cho 10)
unit_digit = product % 10

# In kết quả
print(f"Hàng đơn vị của {a} * {b} là: {unit_digit}")
```
