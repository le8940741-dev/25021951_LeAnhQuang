```python
#W1A3

n = 4
t = 0
for i in range(1, n+1):
    t += i
print(t)

'''Đoạn code trên tính tổng các số nguyên từ 1 đến n (ở đây n = 4), nên kết quả là:

1+2+3+4=10'''
```

```python
#W1A4

numbers = [1, 2, 3, 4]
for x in numbers:
    if x % 2 == 0:
        print(x, "là số chẵn")
    else:
        print(x, "là số lẻ")


'''Đoạn code duyệt qua danh sách [1, 2, 3, 4], kiểm tra từng số là chẵn hay lẻ, rồi in ra kết quả tương ứng.'''
```
